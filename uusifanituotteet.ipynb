```python
# pip install apyori

# tuo kirjastot
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
# lataa tiedot 
df = pd.read_csv('fanituotteet.csv', sep=',')
df.head()
```

```python
# ota sarakkeiden nimet talteen
colnames = df.columns
colnames
```

```python
df.shape[1]
```

```python
 # Järjestä tapahtumat
transctions = []
for i in range(0, df.shape[0]): #10000, rivit
    transact = []
    for j in range(1,df.shape[1]): #21, sarakkeet ei ID
        if df.values[i,j]==1:
            transact.append(colnames[j])
            #print(transaction)
    transctions.append(transact)
transctions
print(transctions[0:10]) #tulostetaan 10 ensimmäistä riviä
```

```python
# mallinnetaan apriori-algoritmillä ja tuodaan tulokset listaaan
# Kokeile säätää parametrejä
results = list(apriori(transactions = transctions,  min_support = 0.1, min_confidence = 0.1, min_lift = 1.1, min_length = 1, max_length = 5))
print(results)
```

```python
# Hae mittarit säännöille ja lisää ne listaan
data = []
print("Supp.  Conf.  Rule")
print(80*"-")
for a in results: # tulokset
    support = a.support
    items = a.items    
    rules = a.ordered_statistics
    print(rules)
    for b in rules: #säännöt
        lhs = b.items_base #left hand side
        rhs = b.items_add # right hand side
        conf = b.confidence
        lift = b.lift
        data.append([support, conf, lhs, rhs, lift])
        print("%6.4f %6.4f %s ---> %s %f" % (support, conf, lhs, rhs, lift))
        print("")
len(data)
```

```python
columns = ["Support", "Confidence", "LHS", "RHS", "Lift"]
data1 = pd.DataFrame(data, columns=columns)
data1.head(15)
```

```python
# n, rivien määrä
data1.nlargest(n = 10, columns = 'Confidence')
```

```python
data1.nlargest(n = 10, columns = 'Lift') 
```

```python
df = df.drop('ID', axis= 1)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Extracting the most frequest itemsets via Mlxtend.
# The length column has been added to increase ease of filtering.
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
# printing the frequent itemset
frequent_itemsets.nlargest(10, 'length')
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2, num_itemsets=100000)
rules
```

```python
rules.nlargest(10, 'confidence')
```
